```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.stats import linregress as LR
import pandas as pd
from pandas.util.testing import assert_frame_equal
from sklearn.model_selection import train_test_split
pd.set_option("display.max_columns",200)
```

```python
#make a chronic patient table for this subset of conditions

# read in that pre-made table
result = pd.read_csv('data_files/result_original.csv')

# Note the information in this csv comes from https://www.ccwdata.org/web/guest/condition-categories
conditions = pd.read_csv("data_files/pain_back_conditions.csv", sep=',', dtype = {'code':str,'name':str})

# Add a column called 'chronic' that is a 1 if the patient was diagnosed primarily with a chronic condition, and 0 otherwise

def chronic(df):
    if df['DIAG1'] in list(conditions['code']):
        return 1
    else:
        return 0

result['chronic'] = result.apply(chronic,axis=1)
chronic = result

# Store the df of patients withy chronic conditions for later analysis
# chronic.to_csv("data_files/chronic_patients.csv")
```

```python
# Make sure we can validate our results - we'll do test/train splits, etc 
# on data up to 2016, then validate on 2016

# First we need ot make sure everything is numerical for it to work
df = chronic[['VMONTH', 'AGE', 'SEX', 'RACE', 'ETHNIC', 'RFV1', 'DIAG1', 'MED1', 'PAYTYPE','YEAR','PATWT', 'received','chronic']]
# for ease of use, I'm dropping all 'V...' well-visit diagnoses
def f(x):
    if str(x)[0] == 'V':
        return np.nan
    else:
        return x
    
# for ease of use, I'm dropping all '...-' and '...--' diagnoses
def g(x):
    if str(x)[-2] == '-':
        return str(x)[:-2]
    elif str(x)[-1] == '-':
        return str(x)[:-1]
    else:
        return x
    
df['DIAG1'] =df['DIAG1'].apply(f)
df['DIAG1'] =df['DIAG1'].apply(g)

# Fill nans in age columns with average of ages
df["AGE"] = df["AGE"].fillna(df["AGE"].mean())

# fill paytype with -8 for unknown
df["PAYTYPE"] = df["PAYTYPE"].fillna(-8)

df = df.dropna(axis = 0, how='any')

#this sets the diagnosis row to floats. Any leading 0s are now gone...
df.DIAG1 = df.DIAG1.astype(np.float64)

df
```

```python
#make histograms showing ages of chronic patients receiving opiates vs all patients receiving opiates

opiates = df[df.received == 1]
chronic_opiates = opiates[opiates.chronic == 1]

#must multiply visits by patient weight (patwt)
op_wts = np.array(opiates.PATWT)
chronic_wts = np.array(chronic_opiates.PATWT)

chronic_ages = np.array(chronic_opiates.AGE)

plt.hist(chronic_ages, weights=chronic_wts, bins=10)
# plt.title('Chronic condition patients receiving opiates ages')
plt.show()
```

```python
#make line chart showing chronic patients vs all patients over time

year_data_tot= []
year_data_chron= []
chron = df[df.chronic == 1]
for year in range(1993, 2016,2):
    col = 'RACE'
    tmp_total = chronic.loc[(chronic['YEAR'] == year) | (chronic['YEAR'] == year+1)]
    tmp_total = tmp_total[tmp_total[col].notnull()]
    tot_wts = np.sum(tmp_total.PATWT)
    
    tmp_chron = chron.loc[(chron['YEAR'] == year) | (chron['YEAR'] == year+1)]
    tmp_chron = tmp_chron[tmp_chron[col].notnull()]
    op_wts = np.sum(tmp_chron.PATWT)
    
    year_data_tot.append(op_wts/tot_wts*100)

years = np.arange(1993, 2016,2)
plt.plot(years, year_data_tot, label="total")
#% total patients who have malignant cancer
# plt.title("Percentage of Total Patients with malignant cancer, per 2 Years, Weighted")
# plt.show()

plt.xlabel("year")
plt.ylabel("percentage of opiates")
plt.legend()
plt.show()
```

```python
only_opiates = chronic[chronic['received'] == 1] #only if they received opiates
only_opiates = only_opiates[only_opiates['chronic'] == 1] #only if they are chronic patients
only_opiates.plot(kind="Hexbin", x="AGE", y="YEAR",C='PATWT',reduce_C_function=np.sum, cmap='coolwarm', gridsize=15,sharex=False)#title='Distribution of Chronic patients Receiving Opiates',

#Labeling Colobar
label = plt.gcf()
cax = label.get_axes()[1]
cax.set_ylabel('Number of People')
plt.show()


average_age = only_opiates.groupby("YEAR")
holder = []
yrs = []
for k, yr in average_age:
    yrs.append(k)
    holder.append(sum(yr['AGE'].values*yr['PATWT'].values)/ sum(yr['PATWT'].values))
   
yrs = np.array(yrs)
slope, intercept, r_value, p_value, stdrr = LR(yrs, holder)
plt.scatter(yrs, holder, color='b', label="Data")
plt.plot(yrs, yrs*slope+intercept, color='r', label="Linear Regression")
# plt.title("Average Age of Chronic patients Receiveing Opiates")
plt.xlabel("Year")
plt.ylabel("Average Age")
plt.legend()
plt.show()
```
