```python
# Note the information in this csv comes from https://www.ccwdata.org/web/guest/condition-categories
chronic = pd.read_csv("data_files/conditions.csv", sep=',', dtype = {'code':str,'name':str})
pain = pd.read_csv("data_files/pain_all_conditions.csv", sep=',', dtype = {'code':str,'name':str})
benign_cancer = pd.read_csv("data_files/benign_cancer_conditions.csv", sep=',', dtype = {'code':str,'name':str})
headache = pd.read_csv("data_files/pain_headache_conditions.csv", sep=',', dtype = {'code':str,'name':str})
heart_lung = pd.read_csv("data_files/heart_lung_conditions.csv", sep=',', dtype = {'code':str,'name':str})
heart= pd.read_csv("data_files/heart_conditions.csv", sep=',', dtype = {'code':str,'name':str})
joint_osteo = pd.read_csv("data_files/pain_joint_neck_back_osteo.csv", sep=',', dtype = {'code':str,'name':str})
lung = pd.read_csv("data_files/lung_conditions.csv", sep=',', dtype = {'code':str,'name':str})
malignant_cancer = pd.read_csv("data_files/malignant_cancer_conditions.csv", sep=',', dtype = {'code':str,'name':str})
neck_back = pd.read_csv("data_files/pain_neck_back_conditions.csv", sep=',', dtype = {'code':str,'name':str})
other_joint = pd.read_csv("data_files/pain_joint_other_conditions.csv", sep=',', dtype = {'code':str,'name':str})
other_pain = pd.read_csv("data_files/pain_other_conditions.csv", sep=',', dtype = {'code':str,'name':str})


# Add a column called 'chronic' that is a 1 if the patient was diagnosed primarily with a chronic condition, and 0 otherwise

def chronic_func(df):
    if df['DIAG1'] in list(chronic['code']):
        return 1
    elif df['DIAG2'] in list(chronic['code']):
        return 1
    elif df['DIAG3'] in list(chronic['code']):
        return 1
    elif df['DIAG4'] in list(chronic['code']):
        return 1
    elif df['DIAG5'] in list(chronic['code']):
        return 1
    else:
        return 0
def pain_func(df):
    if df['DIAG1'] in list(pain['code']):
        return 1
    elif df['DIAG2'] in list(pain['code']):
        return 1
    elif df['DIAG3'] in list(pain['code']):
        return 1
    elif df['DIAG4'] in list(pain['code']):
        return 1
    elif df['DIAG5'] in list(pain['code']):
        return 1
    else:
        return 0
def benign_cancer_func(df):
    if df['DIAG1'] in list(benign_cancer['code']):
        return 1
    elif df['DIAG2'] in list(benign_cancer['code']):
        return 1
    elif df['DIAG3'] in list(benign_cancer['code']):
        return 1
    elif df['DIAG4'] in list(benign_cancer['code']):
        return 1
    elif df['DIAG5'] in list(benign_cancer['code']):
        return 1
    else:
        return 0
def headache_func(df):
    if df['DIAG1'] in list(headache['code']):
        return 1
    elif df['DIAG2'] in list(headache['code']):
        return 1
    elif df['DIAG3'] in list(headache['code']):
        return 1
    elif df['DIAG4'] in list(headache['code']):
        return 1
    elif df['DIAG5'] in list(headache['code']):
        return 1
    else:
        return 0
def heart_lung_func(df):
    if df['DIAG1'] in list(heart_lung['code']):
        return 1
    elif df['DIAG2'] in list(heart_lung['code']):
        return 1
    elif df['DIAG3'] in list(heart_lung['code']):
        return 1
    elif df['DIAG4'] in list(heart_lung['code']):
        return 1
    elif df['DIAG5'] in list(heart_lung['code']):
        return 1
    else:
        return 0
def heart_func(df):
    if df['DIAG1'] in list(heart['code']):
        return 1
    elif df['DIAG2'] in list(heart['code']):
        return 1
    elif df['DIAG3'] in list(heart['code']):
        return 1
    elif df['DIAG4'] in list(heart['code']):
        return 1
    elif df['DIAG5'] in list(heart['code']):
        return 1
    else:
        return 0
def joint_osteo_func(df):
    if df['DIAG1'] in list(joint_osteo['code']):
        return 1
    elif df['DIAG2'] in list(joint_osteo['code']):
        return 1
    elif df['DIAG3'] in list(joint_osteo['code']):
        return 1
    elif df['DIAG4'] in list(joint_osteo['code']):
        return 1
    elif df['DIAG5'] in list(joint_osteo['code']):
        return 1
    else:
        return 0
def lung_func(df):
    if df['DIAG1'] in list(lung['code']):
        return 1
    elif df['DIAG2'] in list(lung['code']):
        return 1
    elif df['DIAG3'] in list(lung['code']):
        return 1
    elif df['DIAG4'] in list(lung['code']):
        return 1
    elif df['DIAG5'] in list(lung['code']):
        return 1
    else:
        return 0
def malignant_cancer_func(df):
    if df['DIAG1'] in list(malignant_cancer['code']):
        return 1
    elif df['DIAG2'] in list(malignant_cancer['code']):
        return 1
    elif df['DIAG3'] in list(malignant_cancer['code']):
        return 1
    elif df['DIAG4'] in list(malignant_cancer['code']):
        return 1
    elif df['DIAG5'] in list(malignant_cancer['code']):
        return 1
    else:
        return 0
def neck_back_func(df):
    if df['DIAG1'] in list(neck_back['code']):
        return 1
    elif df['DIAG2'] in list(neck_back['code']):
        return 1
    elif df['DIAG3'] in list(neck_back['code']):
        return 1
    elif df['DIAG4'] in list(neck_back['code']):
        return 1
    elif df['DIAG5'] in list(neck_back['code']):
        return 1
    else:
        return 0
def other_joint_func(df):
    if df['DIAG1'] in list(other_joint['code']):
        return 1
    elif df['DIAG2'] in list(other_joint['code']):
        return 1
    elif df['DIAG3'] in list(other_joint['code']):
        return 1
    elif df['DIAG4'] in list(other_joint['code']):
        return 1
    elif df['DIAG5'] in list(other_joint['code']):
        return 1
    else:
        return 0
def other_pain_func(df):
    if df['DIAG1'] in list(other_pain['code']):
        return 1
    elif df['DIAG2'] in list(other_pain['code']):
        return 1
    elif df['DIAG3'] in list(other_pain['code']):
        return 1
    elif df['DIAG4'] in list(other_pain['code']):
        return 1
    elif df['DIAG5'] in list(other_pain['code']):
        return 1
    else:
        return 0

result['chronic'] = result.apply(chronic_func,axis=1)
result['pain'] = result.apply(pain_func,axis=1)
result['benign_cancer'] = result.apply(benign_cancer_func,axis=1)
result['headache'] = result.apply(headache_func,axis=1)
result['heart_lung'] = result.apply(heart_lung_func,axis=1)
result['heart'] = result.apply(heart_func,axis=1)
result['joint_osteo'] = result.apply(joint_osteo_func,axis=1)
result['lung'] = result.apply(lung_func,axis=1)
result['malignant_cancer'] = result.apply(malignant_cancer_func,axis=1)
result['neck_back'] = result.apply(neck_back_func,axis=1)
result['other_joint'] = result.apply(other_joint_func,axis=1)
result['other_pain'] = result.apply(other_pain_func,axis=1)

chronic = result
chronic
```

```python
#Make sure we can validate our results - we'll do test/train splits, etc 
# on data up to 2016, then validate on 2016

# First we need ot make sure everything is numerical for it to work
df = chronic[['VMONTH', 'AGE', 'SEX', 'RACE', 'ETHNIC', 'RFV1', 'DIAG1','DIAG2','DIAG3','DIAG4','DIAG5',
              'MED1', 'PAYTYPE','YEAR','PATWT', 'received','chronic','pain','benign_cancer','headache',
              'heart_lung','heart','joint_osteo','lung','malignant_cancer','neck_back','other_joint','other_pain']]
# for ease of use, I'm dropping all 'V...' well-visit diagnoses
def f(x):
    if str(x)[0] == 'V':
        return np.nan
    else:
        return x
    
# for ease of use, I'm truncating all '...-' and '...--' diagnoses
def g(x):
    if str(x)[-2] == '-':
        return str(x)[:-2]
    elif str(x)[-1] == '-':
        return str(x)[:-1]
    else:
        return x
    

for i in range(5):
    df[f'DIAG{i+1}'] =df[f'DIAG{i+1}'].apply(f)
    df[f'DIAG{i+1}'] =df[f'DIAG{i+1}'].apply(g)

# Fill nans in age columns with average of ages
df["AGE"] = df["AGE"].fillna(df["AGE"].mean())

# fill paytype with -8 for unknown
df["PAYTYPE"] = df["PAYTYPE"].fillna(-8)

df = df.dropna(axis = 0, subset = ['VMONTH', 'AGE', 'SEX', 'RACE', 'DIAG1'],how='any')

#this sets the diagnosis row to floats. Any leading 0s are now gone. We first set NaN vals to 0 (not a valid diagnosis)
df["DIAG2"] = df["DIAG2"].fillna(0)
df["DIAG3"] = df["DIAG3"].fillna(0)
df["DIAG4"] = df["DIAG4"].fillna(0)
df["DIAG5"] = df["DIAG5"].fillna(0)

df.DIAG1 = df.DIAG1.astype(np.int)
df.DIAG2 = df.DIAG2.astype(np.int)
df.DIAG3 = df.DIAG3.astype(np.int)
df.DIAG4 = df.DIAG4.astype(np.int)
df.DIAG5 = df.DIAG5.astype(np.int)

df
```
